```python
url = "https://lib.stat.cmu.edu/datasets/boston"

# Step 1: Load raw data
raw_df = pd.read_csv(url, sep="\s+", skiprows=22, header=None)

# Step 2: Split rows (dataset is split across lines)
data_part1 = raw_df.iloc[::2, :]   # even rows
data_part2 = raw_df.iloc[1::2, :]  # odd rows

# Step 3: Combine into proper structure
X = np.hstack([data_part1.values, data_part2.values[:, :2]])
y = data_part2.values[:, 2]

# Step 4: Column names
columns = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
]

# Step 5: Create DataFrame
df = pd.DataFrame(np.hstack([X, y.reshape(-1, 1)]), columns=columns)

# Step 6: Preview
print(df.head())
print(df.shape)
```

Preparing Dataset

```python
from sklearn.datasets import fetch_openml

boston = fetch_openml(name="boston", version=1)

dataset = pd.DataFrame(boston.data, columns = boston.feature_names)
target = boston.target
dataset.head()
```

```python
dataset['Price']=boston.target
dataset.head()
```

```python
# Independent and Dependent

X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]

X.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

Model Training

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
# print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# which parameters the model has been trained

regression.get_params()
```

```python
# Prediction with test data

reg_pred = regression.predict(X_test)
reg_pred
```

```python
# plot a scatter for the  prediction
plt.scatter(y_test,reg_pred)
```

```python
# Residuals

residuals = y_test - reg_pred
residuals
```

```python
#plot this residuals

sns.displot(residuals,kind='kde')
```

```python
# Scatter plot with respect to prediction and residuals
# Uniform distribution

plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))

print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
# R Squre and adjusted R Square

from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
# display adjusted R-squared

1- (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
